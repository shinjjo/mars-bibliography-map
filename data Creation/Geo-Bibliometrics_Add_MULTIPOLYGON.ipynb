```python
#Source from 
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
import pandas as pd
import ogr
from matplotlib import pylab
import json
import csv
```

```python
class polygon_extract:
    def __init__(self, id):
        self.name = id
        self.urlIAU = "https://planetarynames.wr.usgs.gov/Feature/"
        self.url = self.urlIAU + self.name
        print("url to call: " + self.url)
    
    def extract(self):
        self.r = requests.get(self.url)
        self.data = self.r.text
        self.soup = BeautifulSoup(self.data, "lxml")
        self.feature_panel = self.soup.findAll('table', {'class': 'feature_detail'})
        
        for feature in self.feature_panel:
            try:
                self.name = feature.find('h2').text
            except AttributeError:
                continue
        self.list1 = self.soup.body.findAll(text = re.compile('MULTIPOLYGON'))        
        self.str1 = ''.join(self.list1)
        self.splitList = self.str1.split('\n')
        for item in self.splitList:
            if item.find("MULTIPOLYGON") != -1:
                self.geomStr = item.split('\'')[1]
                self.geomOGR = ogr.CreateGeometryFromWkt(self.geomStr)
                self.coords = json.loads(self.geomOGR.ExportToJson())['coordinates'][0]
                for i in self.coords:
                    for j in i:
                        if j[0] > 180 or j[0] < -180:
                            j[0] = j[0]-360
                        else:
                            pass
                return self.coords
    
            else:
                pass
        return []
```

```python
polygon_extract('38').extract()
```
